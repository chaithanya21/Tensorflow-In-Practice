```
# More Concrete Example with Padding and other words Which arent there in the Vocab

training_data=['Sunil Gavaskar is the first batsman to score ten thousand runs','Virat is one of the fastest batsman to reach ten thousand runs']

```

```
TOKINIZER=Tokenizer(num_words=100 , oov_token='<OOV>')
```

```
TOKINIZER.fit_on_texts(training_data)
word_index=TOKINIZER.word_index
print(word_index)
```

```
seq=TOKINIZER.texts_to_sequences(training_data)
print(seq)
```

```
test_data=['ABD is the greatest entertainer in cricket']
```

```
results=TOKINIZER.texts_to_sequences(test_data)
```

```
print(results)
```

```
from tensorflow.keras.preprocessing.sequence import  pad_sequences
```

```
padded_seq=pad_sequences(seq)
```

```
print(padded_seq)
```

```
# Sarcasm Detection using NLP
from google.colab import files
```

```
import pandas as pd
```

```
Dataframe=pd.read_json('Sarcasm_Headlines_Dataset_v2.json',lines=True)
```

```
Dataframe.head()
```

```
Dataframe.drop(columns=['article_link'],axis=1)
```

```
Sentences=Dataframe['headline']
```

```
Sentences
```

```
Sentences_list=pd.Series.to_list(Sentences)
```

```
Sentences_list
```

```
List_tokenizer=Tokenizer(num_words=1000,oov_token='<oov>')
```

```
List_tokenizer.fit_on_texts(Sentences_list)
word_index=List_tokenizer.word_index
print('word index:',word_index)
Sequences=List_tokenizer.texts_to_sequences(Sentences_list)
print('Sequences:',Sequences)
```

```
word_count=List_tokenizer.word_counts
print(word_count)
```

```
padding=pad_sequences(Sequences,maxlen=10)
```
